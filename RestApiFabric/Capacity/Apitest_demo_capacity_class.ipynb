### Create, Delete, Update Capacity Class

```python
import requests
from typing import Optional

class AzureAPIClient:
    def __init__(self, subscription_id: str, access_token: str):
        self.subscription_id = subscription_id
        self.access_token = access_token
        self.base_url = "https://management.azure.com"

    def delete_capacity(self, resource_group_name: str, provider_name: str, resource_name: str, api_version: str) -> Optional[requests.Response]:
        try:
            # Construct the URL using the instance variables and method parameters
            url = f"{self.base_url}/subscriptions/{self.subscription_id}/resourceGroups/{resource_group_name}/providers/{provider_name}/capacities/{resource_name}?api-version={api_version}"

            # Define headers
            headers = {
                'Authorization': f'Bearer {self.access_token}',
                'Content-Type': 'application/json'
            }
            
            # Make the DELETE request
            response = requests.delete(url, headers=headers)

            # Check for HTTP errors
            response.raise_for_status()

            # Return the response
            return response

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")  # Handle specific HTTP errors
        except requests.exceptions.RequestException as req_err:
            print(f"Request error occurred: {req_err}")  # Handle other request exceptions
        except Exception as err:
            print(f"An error occurred: {err}")  # Handle any other exceptions

        return None  # Return None if an exception occurs

        
    def pause_capacity(self, resource_group_name: str, provider_name: str, resource_name: str, api_version: str) -> Optional[requests.Response]:
        try:
            # Construct the URL for pausing the capacity
            url = f"{self.base_url}/subscriptions/{self.subscription_id}/resourceGroups/{resource_group_name}/providers/{provider_name}/capacities/{resource_name}/suspend?api-version={api_version}"

            # Define headers
            headers = {
                'Authorization': f'Bearer {self.access_token}',
                'Content-Type': 'application/json'
            }

            # Make the POST request to pause the capacity
            response = requests.post(url, headers=headers)

            # Check for HTTP errors
            response.raise_for_status()

            # Return the response
            return response

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")  # Handle specific HTTP errors
        except requests.exceptions.RequestException as req_err:
            print(f"Request error occurred: {req_err}")  # Handle other request exceptions
        except Exception as err:
            print(f"An error occurred: {err}")  # Handle any other exceptions

        return None  # Return None if an exception occurs

          def create_capacity(self, resource_group_name: str, provider_name: str, resource_name: str, api_version: str, sku_name: str, sku_tier: str, tag_key: str, tag_value: str, location: str, admin_email: str) -> Optional[requests.Response]:
        try:
            #base_url = "https://management.azure.com"
            # Construct the URL using the instance variables and method parameters
            url = f"{self.base_url}/subscriptions/{self.subscription_id}/resourceGroups/{resource_group_name}/providers/{provider_name}/capacities/{resource_name}?api-version={api_version}"

            print (url)
            # Construct the payload using the variables
            payload = {
                "sku": {
                    "name": sku_name,
                    "tier": sku_tier
                },
                "tags": {
                    tag_key: tag_value
                },
                "location": location,
                "properties": {
                    "administration": {
                        "members": [
                            admin_email
                        ]
                    }
                }
            }

            # Define headers
            headers = {
                'Authorization': f'Bearer {self.access_token}',
                'Content-Type': 'application/json'
            }

            # Make the PUT request
            response = requests.put(url, headers=headers, json=payload)

            # Check for HTTP errors
            response.raise_for_status()

            # Return the response
            return response

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")  # Handle specific HTTP errors
        except requests.exceptions.RequestException as req_err:
            print(f"Request error occurred: {req_err}")  # Handle other request exceptions
        except Exception as err:
            print(f"An error occurred: {err}")  # Handle any other exceptions

        return None  # Return None if an exception occurs

def update_capacity(self, resource_group_name: str, provider_name: str, resource_name: str, api_version: str, sku_name: str, sku_tier: str, tag_key: str, tag_value: str, admin_emails: list) -> Optional[requests.Response]:
        try:
            # Construct the URL for updating the capacity
            url = f"{self.base_url}/subscriptions/{self.subscription_id}/resourceGroups/{resource_group_name}/providers/{provider_name}/capacities/{resource_name}?api-version={api_version}"

            # Construct the payload using the variables
            payload = {
                "sku": {
                    "name": sku_name,
                    "tier": sku_tier
                },
                "tags": {
                    tag_key: tag_value
                },
                "properties": {
                    "administration": {
                        "members": admin_emails
                    }
                }
            }

            # Define headers
            headers = {
                'Authorization': f'Bearer {self.access_token}',
                'Content-Type': 'application/json'
            }

            # Make the PATCH request
            response = requests.patch(url, headers=headers, json=payload)

            # Check for HTTP errors
            response.raise_for_status()

            # Return the response
            return response

        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")  # Handle specific HTTP errors
        except requests.exceptions.RequestException as req_err:
            print(f"Request error occurred: {req_err}")  # Handle other request exceptions
        except Exception as err:
            print(f"An error occurred: {err}")  # Handle any other exceptions

        return None  # Return None if an exception occurs


    def capacity_exists(self, resource_name: str, provider_name: str) -> bool:
        try:
            url = f"{self.base_url}/subscriptions/{self.subscription_id}/providers/{provider_name}/capacities?api-version=2021-01-01"
            headers = {
                'Authorization': f'Bearer {self.access_token}',
                'Content-Type': 'application/json'
            }
            response = requests.get(url, headers=headers)
            response.raise_for_status()
            capacities = response.json().get('value', [])
            for capacity in capacities:
                if capacity['name'] == resource_name:
                    return True
            return False
        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")
        except requests.exceptions.RequestException as req_err:
            print(f"Request error occurred: {req_err}")
        except Exception as err:
            print(f"An error occurred: {err}")
        return False
```
